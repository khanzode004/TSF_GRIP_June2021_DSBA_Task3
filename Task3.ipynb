### Removing an Unwanted Feature

### Data Visualization

```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category', data=df)
plt.show()
```

```python
print(df1['State'].value_counts())
plt.figure(figsize=(15,8))
sns.countplot(x=df1['State'])
plt.xticks(rotation=90)
plt.show()
```

### Correlation between Variables and its Heatmap

```python
df.corr()
```

```python
fig,axes = plt.subplots(1,1,figsize=(9,6))
sns.heatmap(df.corr(), annot= True)
plt.show()
```

### Contribution of each Sub-Category or products towards Profits

```python
plt.figure(figsize=(40,25))
sns.barplot(x=df['Sub-Category'], y=df['Profit'])
sns.set(font_scale=7)
plt.xticks(rotation='vertical')
```

### Profit vs Discount Line Graph

```python
plt.figure(figsize = (10,4))
sns.lineplot('Discount', 'Profit', data = df, color = 'r', label= 'Discount')
plt.xticks(rotation='vertical')
sns.set(font_scale=2)
plt.legend()
```

### Top 20 Aggregate Profit for each possible combination of State, Category and Sub-category

```python
grouped=pd.DataFrame(df.groupby(['Ship Mode','Segment','Category','Sub-Category','State','Region'])['Quantity','Discount','Sales','Profit'].sum().reset_index())
grouped.sort_values(by='Profit', ascending=False, inplace = True)
grouped.head(20)
```

### Plotting top states with highest mean profit¶


```python
statewise_profit = df.groupby("State").Profit.agg(["mean"])
statewise_profit.sort_values(by='mean', ascending=False, inplace=True)
```

```python
statewise_profit
```

```python
statewise_profit['mean'].head(10).plot(kind='bar')
plt.xticks(rotation='vertical')
```

### Profit vs Sales Scatter Plot
